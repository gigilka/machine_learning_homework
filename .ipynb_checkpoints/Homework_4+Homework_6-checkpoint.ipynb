# Описание задачи

Датасет содержит 236378 записей о пациентах, имеющих предиабетный или диабетный статус и не имеющих диабета. Задача: по имеющимся признакам (таким как имт, уровень холестерина, образ жизни и привычки, доходы, образования, пол и др.) предсказать, к какому классу относится пациент: имеющий предиабет или диабет, или не имеющий.

# Нормализация признаков

```python
# метод стандартизации
features = dataset.drop('Diabetes_binary', axis=1)
target = dataset['Diabetes_binary']

features_standardized = (features - features.mean(axis=0)) / features.std(axis=0)
dsn = features_standardized.join(target)
```

# Классификатор

```python
X = dsn.drop('Diabetes_binary', axis=1)
y = dsn['Diabetes_binary'].astype(int) 

# Разделение на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Вывод размерности полученных выборок
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy of KNN classifier on test set: {accuracy:.2%}')


conf_matrix = sklearn.metrics.confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='g')
plt.title('Confusion Matrix')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=10)

knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy of KNN classifier on test set: {accuracy:.2%}')

conf_matrix = sklearn.metrics.confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='g')
plt.title('Confusion Matrix')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```

```python
knn = KNeighborsClassifier()
param_grid = {'n_neighbors': list(range(1, 11))}
grid_search = GridSearchCV(knn, param_grid, cv=5) 
grid_search.fit(X_train, y_train)
print(f'Лучшее значение параметра: {grid_search.best_params_}')
print(f'Лучшая оценка модели: {grid_search.best_score_}')
best_knn = grid_search.best_estimator_
y_pred = best_knn.predict(X_test)
from sklearn.metrics import classification_report, accuracy_score
print(f'Точность лучшей модели k-NN на тестовых данных: {accuracy_score(y_test, y_pred)}')
print(classification_report(y_test, y_pred))
```

# Выводы

- Большее количество k-соседей имеет очень малое влияние на улучшение точности
- Модель достигает меньшей производительности на распознавании менее распространенного класса (диабет и предиабет), но хорошо распознаёт недиабет.

# Домашняя работа №6

### Логистическая регрессия

```python
lr = LogisticRegression(random_state=42, max_iter=1000)
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test) 
conf_matrix = sklearn.metrics.confusion_matrix(y_test, y_pred)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy of Logic regression: {accuracy:.2%}')

sns.heatmap(conf_matrix, annot=True, fmt='g')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

### Random forest

```python
rf = RandomForestClassifier(random_state=42, n_estimators = 1000)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy of Random Forest classifier on test set: {accuracy:.2%}')

conf_matrix = sklearn.metrics.confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='g')
plt.title('Confusion Matrix for Random Forest')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```

### Полносвязная нейронная сеть с 1 скрытым слоем

```python
model = Sequential()
model.add(Dense(50, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(50, activation='relu')) 
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5).astype(int)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy of the neural network on test set: {accuracy:.2%}')

conf_matrix = sklearn.metrics.confusion_matrix(y_test, y_pred)

plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='g')
plt.title('Confusion Matrix')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```

### Подбор гиперпараметров для логистической регрессии

```python
param_grid = [
  {
    'C': np.logspace(-4, 4, 20),
    'solver': ['lbfgs', 'newton-cg', 'sag', 'liblinear'],
    'penalty': ['l2']
  }
]

grid_search = GridSearchCV(
    estimator=LogisticRegression(random_state=42, max_iter=1000),
    param_grid=param_grid,
    scoring='accuracy',
    cv=3,
    verbose=1,
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_estimator = grid_search.best_estimator_
print('Best:', best_params)

y_train_pred = best_estimator.predict(X_train)
train_accuracy = accuracy_score(y_train, y_train_pred)
print(f'Accuracy of Logistic Regression with optimized hyperparameters on train: {train_accuracy:.2%}')

y_pred = best_estimator.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy of Logistic Regression with optimized hyperparameters on test: {accuracy:.2%}')

conf_matrix = sklearn.metrics.confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='g')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

results = pd.DataFrame(grid_search.cv_results_)

plt.figure(figsize=(10, 8))
Cs = results['param_C'].to_numpy().astype(float)
plt.plot(Cs, 1 - results['mean_test_score'], 'o-')
plt.title('Error vs C')
plt.xlabel('C')
plt.ylabel('Mean error')
plt.show()

plt.figure(figsize=(10, 8))
Cs = results['param_C'].to_numpy().astype(float)
plt.plot(np.log10(Cs), 1 - results['mean_test_score'], 'o-')
plt.title('Error vs C (on log scale)')
plt.xlabel('Log10 of C')
plt.ylabel('Mean error')
plt.show()

plt.figure(figsize=(10, 8))
solvers = results['param_solver'].to_numpy()
for solver in np.unique(solvers):
    solver_mask = (solvers == solver)
    plt.plot(np.log10(Cs[solver_mask]), 1 - results['mean_test_score'][solver_mask], 'o-', label=solver)
plt.title('Error vs Solver')
plt.xlabel('Log10 of C')
plt.ylabel('Mean error')
plt.legend()
plt.show() 

plt.figure(figsize=(10, 8))
solvers = results['param_solver'].to_numpy()
for solver in np.unique(solvers):
    solver_mask = (solvers == solver)
    plt.plot(Cs[solver_mask], 1 - results['mean_test_score'][solver_mask], 'o-', label=solver)
plt.title('Error vs Solver and C')
plt.xlabel('C')
plt.ylabel('Mean error')
plt.legend()
plt.show()
```

# Выводы

Для зафиксированной функции потерь (L2-регуляризация):\
При значении параметра С, близкому к 0.00026 (log_10(0.00026) ≈ -3.58), достигается минимум (точный минимум = 0.00026366508987303583). \
В большинстве случаев выигрывает метод координатного спуска (liblinear), но выигрыш незначительный, достигается только при оптимальном С и уменьшается при увеличении С (уход от оптимальности) - ошибки всех методов становятся одинаковыми.\
Маленькие значения C приводят к более высокой регуляризации (больше обобщений), в то время как большие значения C снижают регуляризацию и делают модель более подогнанной под обучающие данные.\
Также при увеличении С ошибка для всех методов логистической регрессии уменьшается, затем увеличивается - это также указывает на переобучение модели при высоких значениях C.\
Но точность на обучающей и тестовый выборках почти совпадает (85.47%  85.41), это указывает, что модель не слишком переобучена или недообучена.

Из всего можно сделать вывод: однозначно оценить переобучение или недообучение модели нельзя; в общем случае модель "стабильна", но при некоторых значениях параметров может наблюдаться переобучение.
\

Из confusion matrix:
Все методы (лог. регр., rand om forest, нейронная сеть со скрытыми слоями) хорошо предсказывают наиболее распространённый класс, а наименее расп. предсказывают достаточно плохо. Это, вероятно, связано с несбалансированными классами. Улучшением может быть применение соответствующих методов балансировки классов, такие как взвешивание классов или пересэмплирование
